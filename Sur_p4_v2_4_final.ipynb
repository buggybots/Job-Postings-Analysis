```python
def sal_cat(sal_string):
    if 'year' in sal_string.lower():
        return 'Annual'
    elif 'month' in sal_string.lower():
        return 'Monthly'
    elif 'day' in sal_string.lower():
        return 'Daily'
    elif 'hour' in sal_string.lower():
        return 'Hourly'
    elif 'week' in sal_string.lower():
        return 'Weekly'
    else:
        return 'other'
```

```python
# function takes in a salary as a string, removes any text, then splits the numbers into a list, 
# then returns either the average of two numbers or the single number itself.
def sal_avg(sal_string):
    sal_range = ''.join(re.findall(r"([\d-]+)", sal_string))
    sals_list = sal_range.split('-')
    if len(sals_list)==2:
        return int(np.mean([int(sals_list[0]), int(sals_list[1])]))
    else:
        return int(sals_list[0])
```

```python
df_annual['avg_salary'] = df_annual['salary'].apply(sal_avg)
```

```python
def text_cleaner(description):
    """Function takes in the string job description as a parameter, converts the string to lowercase,
    removes any numbers, special characters and email addresses. Returns cleaned job description."""
    
    # Turning text in to lowercase
    cleaned_description = description.lower()
    # removing numbers that are in the middle of the text
    cleaned_description = re.sub('\w*\d\w*', ' ', cleaned_description)
    # taking out some special characters
    cleaned_description = re.sub('[;$,\n.]', ' ', cleaned_description)
    # removing email addresses
    cleaned_description = re.sub('([a-zA-Z0-9._-]+@[a-zA-Z0-9._-]+\.[a-zA-Z0-9_-]+)' , ' ',cleaned_description)
    
    return cleaned_description
```

```python
df_annual['clean_descr'] = df_annual.descr.apply(text_cleaner)
```

## LogisticRegression

```python
def token_logistic_model(vect):
    X_train_dtm = vect.fit_transform(X_train)           
    print(f'Number of Features: {X_train_dtm.shape[1]}')
    
    logreg = LogisticRegressionCV(cv=5, max_iter=500)
    pipeline = Pipeline([('cvec',vect),('logistic',logreg)])
    pipeline.fit(X_train, y_train)
    y_pred_class = pipeline.predict(X_test)
    score = metrics.accuracy_score(y_test,y_pred_class)
    
    print(f'Model accuracy: {score}')
#     print(f'Score: {pipeline.score(X_test, y_test)}')
    print(f'Precision: {precision_score(y_test,y_pred_class)}')
    print(classification_report(y_test, y_pred_class))
    print(confusion_matrix(y_test, y_pred_class))
    
    # calculate the fpr and tpr for all thresholds of the classification
     
    probs = pipeline.predict_proba(X_test)
    preds = probs[:,1]
    
    fpr, tpr, threshold = metrics.roc_curve(y_test, preds)
    roc_auc = metrics.auc(fpr, tpr)
    plt.title('ROC Curve')
    plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
    plt.legend(loc = 'lower right')
    plt.plot([0, 1], [0, 1],'r--')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.show()
```
